```python
# Dispersion: Calculate standard deviation, variance, range, and interquartile range for turnout and margins.
# Turnout

mean_turnout = df['Turnout'].mean()
print(mean_turnout)
diff = df['Turnout']-mean_turnout
diff
square_diff = diff ** 2
square_diff

variance = square_diff.mean()
print(variance)

# var_data = df['Turnout'].var(ddof=0)
# var_data
```

```python
# Margin

mean_margin = df['Margin'].mean()
print(mean_margin)
diff1 = df['Margin'] - mean_margin
diff1
square_diff1 = diff1 ** 2
square_diff1

variance1 = square_diff1.mean()
print(variance1)

# var_data1 = df['Margin'].var(ddof=0)
# var_data1
```

Standard deviation

```python
# Turnout
std_deviation = variance ** (1/2)
print(std_deviation)

# Margin
std_deviation1 = variance1 ** (1/2)
print(std_deviation1)

# std_mar = df['Margin'].std(ddof=0)
# print(std_mar)
```

Margin of Victory:
Probability: Probability of different margins of victory (e.g., close races vs. landslides).

Statistics: Average margin of victory, standard deviation.

Distribution: Distribution of victory margins across constituencies.

```python
# Probability: Probability of different margins of victory (e.g., close races vs. landslides).
def classify_margin(margin_percent):
  if margin_percent <= 5:
    return "race close"
  elif margin_percent > 5 and margin_percent <= 10:
    return 'Moderate Race'
  else:
    return 'Landslide'

df['Margin Category'] = df['Margin %'].apply(classify_margin)

df
# df.drop('Margin Victory', axis=1, inplace=True)
```

Turnout vs. Margin:

1. The correlation coefficient between Turnout and Margin is 0.17.
2. This indicates a weak positive correlation between voter turnout and the winning margin. As turnout increases, the margin tends to increase slightly.


Turnout vs. Margin %:

1. The correlation coefficient between Turnout and Margin % is 0.03.
2. This indicates a very weak positive correlation (almost negligible) between voter turnout and the margin percentage. There's hardly any relationship between the turnout and the margin percentage.

Margin vs. Margin %:

1. The correlation coefficient between Margin and Margin % is 0.92.
2. This indicates a very strong positive correlation between the margin and the margin percentage. As the absolute winning margin increases, the margin percentage also increases significantly. This is expected as a higher number of votes won will directly impact the percentage margin as well.


#Regression Analysis


```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Define the feature and target variables
X_turnout = df[['Turnout']]
y_margin = df['Margin']
y_margin_pct = df['Margin %']
```

```python
X_train_margin, X_test_margin, y_train_margin, y_test_margin = train_test_split(X_turnout, y_margin, test_size=0.2, random_state=42)

# Split the data into training and testing sets for Margin %
X_train_margin_pct, X_test_margin_pct, y_train_margin_pct, y_test_margin_pct = train_test_split(X_turnout, y_margin_pct, test_size=0.2, random_state=42)
```

```python
# Create and train the linear regression model for Margin
model_margin = LinearRegression()
model_margin.fit(X_train_margin, y_train_margin)

# Create and train the linear regression model for Margin %
model_margin_pct = LinearRegression()
model_margin_pct.fit(X_train_margin_pct, y_train_margin_pct)
```

```python
y_pred_margin = model_margin.predict(X_test_margin)
y_pred_margin_pct = model_margin_pct.predict(X_test_margin_pct)
```

```python
y_pred_margin
```

```python

# Plot for Margin
plt.subplot(1, 2, 1)
plt.scatter(X_test_margin, y_test_margin, color='blue', label='Actual')
plt.plot(X_test_margin, y_pred_margin, color='red', linewidth=2, label='Predicted')
plt.title('Linear Regression: Turnout vs Margin')
plt.xlabel('Turnout')
plt.ylabel('Margin')
plt.legend()
```

```python
# Plot for Margin %
plt.subplot(1, 2, 2)
plt.scatter(X_test_margin_pct, y_test_margin_pct, color='blue', label='Actual')
plt.plot(X_test_margin_pct, y_pred_margin_pct, color='red', linewidth=2, label='Predicted')
plt.title('Linear Regression: Turnout vs Margin %')
plt.xlabel('Turnout')
plt.ylabel('Margin %')
plt.legend()
```
